```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from pathlib import Path
import warnings; warnings.filterwarnings("ignore")
```

```python
# Load Titanic dataset
project_root = Path.cwd().parent
base_path = project_root / "Dataset"
df = pd.read_csv(base_path / "unprocessed_titanic.csv")
df.head()
```

```python
# Delete irrelevant features
df.drop(["PassengerId", "Name", "Ticket"], axis = 1, inplace = True)

# Drop Cabin feature, it has a lot af missing values
df.drop("Cabin", axis = 1, inplace = True)
```

```python
# Copy unprocessed dataframe for later analysis
df_a = df.copy()

# Change values for better analysis
df_a["Sex"] = df_a["Sex"].map({"male" : 0, "female" : 1})
df_a["Survived"] = df_a["Survived"].map({0 : "No", 1 : "Yes"})
```

```python
# To categorical
cat = [col for col in df.columns if df[col].dtype == "object"]
dummies = pd.get_dummies(df[cat])
df = (pd.concat([dummies, df], axis = 1)).drop(["Embarked", "Sex"], axis = 1)
df.rename(columns = {"Embarked_C" : "C", "Embarked_Q" : "Q", "Embarked_S" : "S", "Sex_female" : "Female", "Sex_male" : "Male"}, inplace = True)
```

```python
# Use KNN to fill nulls in Age feature
df["Age"] = df["Age"].fillna(0)
test = df[df["Age"] == 0].loc[:, df.columns != "Age"]
age_d = df.drop(np.where(df["Age"] == 0)[0])
x_age_d = age_d.loc[:, df.columns != "Age"]
y_age_d = age_d["Age"]

# Fit KNN
knn = KNeighborsRegressor(n_neighbors = 3).fit(x_age_d, y_age_d)
pred = knn.predict(test)

# Fill Age feature nulls
df.loc[df["Age"] == 0, "Age"] = pred
```

```python
# Fill Fare feature nulls: mean fare in each Pclass
for i in range(1, 4):
    df[df["Fare"] == 0][df["Pclass"] == i]["Fare"] = df[df["Pclass"] == i]["Fare"].mean()
```

```python
# Remove outliers with ZScore
z_score = np.abs(stats.zscore(df))
df = df[(z_score < 3.2).all(axis = 1)]
```

```python
# Feature engineering
df["Family Size"] = df["SibSp"] + df["Parch"]

# Drop those features
df.drop(["SibSp", "Parch"], axis = 1, inplace = True)
```

```python
# Rearrenge features
df = df[np.hstack([df.columns[df.columns != "Survived"],"Survived"])]
```

```python
# Train and test split
x = df.loc[:, df.columns != "Survived"]
y = df["Survived"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# Save the datasets
x_train.to_csv(base_path / "x_train.csv", index = False)
y_train.to_csv(base_path / "y_train.csv", index = False)
x_test.to_csv(base_path / "x_test.csv", index = False)
y_test.to_csv(base_path / "y_test.csv", index = False)
```

```python
# Import ML libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

```python
# Build a primary decision tree model
p_d_tree = DecisionTreeClassifier(random_state = 0)

# Fit the model to the data
p_d_tree.fit(x_train, y_train)

# Get predictions of the model
prim_tree_preds = p_d_tree.predict(x_test)

# Evaluation
accuracy_score(y_test, prim_tree_preds)
```

```python
# Hyper parameter tuning using GridSearchCV
d_tree = DecisionTreeClassifier(random_state = 42)
params = {"min_samples_split" : [2, 4, 5, 10],
          "criterion" : ["gini", "entropy"],
          "min_samples_leaf" : [1, 2, 4, 6, 10],
          "max_depth" : [2, 3, 4, 6, None]}
grid_s = GridSearchCV(d_tree, params, cv = 3, scoring = "neg_mean_squared_error")
grid_s.fit(x_train, y_train);
print(grid_s.best_estimator_)
```

```python
# Final tree
final_tree = grid_s.best_estimator_

# Fit the final model to the data
final_tree.fit(x_train, y_train)

# Get predictions of the final model
tree_preds = final_tree.predict(x_test)

# Evaluation
print("Accuracy: ", accuracy_score(y_test, tree_preds))
print("F1 Score: ", f1_score(y_test, tree_preds))
```

```python
# Write results into results.txt file
with open(project_root / "evaluation_results.txt", "a") as f:
    f.write("Decision Tree Results:\n")
    f.write("Accuracy: " + f"{accuracy_score(y_test, tree_preds):.6f}" + "\n")
    f.write("F1 Score: " + f"{f1_score(y_test, tree_preds):.6f}" + "\n\n")
```
